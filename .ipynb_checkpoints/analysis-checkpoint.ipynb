```python
import pandas as pd

df = pd.read_csv("Film_Dataset.csv")
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df['Unique_ID'] = df['Film_Name'] + "_" + df['Language'] + "_" + df['Category']
```

```python
df['Viewing_Month'] = pd.to_datetime(df['Viewing_Month'])
df_2025 = df[df['Viewing_Month'].dt.year == 2025]
```

```python
//EDA
```

```python
import matplotlib.pyplot as plt

df_2025.groupby("Category")["Number_of_Views"].sum().plot(kind="bar")
plt.title("Views by Category (2025)")
plt.show()
```

```python
df_2025.groupby("Language")["Number_of_Views"].sum().plot(kind="bar")
plt.title("Views by Language (2025)")
plt.show()
```

```python
df['Popularity_Score'] = (
    df['Number_of_Views']*0.6 +
    df['Viewer_Rate']*0.4
)
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_film_data.csv")

st.title("IMovie  December 2025 Marketing Dashboard")

# Category Views
st.subheader("Views by Category")
cat = df.groupby("Category")["Number_of_Views"].sum()
st.bar_chart(cat)

# Language Views
st.subheader("Views by Language")
lang = df.groupby("Language")["Number_of_Views"].sum()
st.bar_chart(lang)

```
